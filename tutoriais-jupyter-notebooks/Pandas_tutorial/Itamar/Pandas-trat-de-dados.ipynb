```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
from pandas.io import sql
```

```python
def ler_dado_ctd(nome_arq):
    
    #Iniciando as variaveis básicas
    vetor_dados = []
    arq_lido = open(nome_arq)
    datas = []
    dados = []
    dados_temp = []
    linha_temp = []
    datas_decimais = []
    #Este bloco de programa tem por objetivo filtrar os dados do arquivo lido, segundo o professor estabeleceu na pratica.
    for item in arq_lido:
        vetor_dados.append(item)
    n_linhas = len(vetor_dados)
    for item in range(0, n_linhas , 1):
            if vetor_dados[item][0] != '#' and vetor_dados[item][0 ] != '*':
                linha = vetor_dados[item].split()
                dados_temp.append(linha)
    num = len(dados_temp)
    #aqui inicia-se o bloco que irá fatiar a lista de dados_temp e separar o que é data de dados.
    print(num)
    dados_array = np.array(dados_temp)
    prof = list(dados_array[:,1])
    return dados_array
```

```python
resultado = ler_dado_ctd('PNB1_003.cnv')
```

```python
def conv_num(lista):
    lista_temp = []
    for i in range(len(lista[0])):
        lista_temp.append([])
        for j in range(len(lista)):
            temp = float(lista[j][i])
            lista_temp[i].append(temp)

    return lista_temp
```

```python
matriz_dados = conv_num(resultado)
matriz_array = np.array(matriz_dados)
matriz_trans = matriz_array.T

print(len(matriz_trans))
```

```python
def posicoes(lista, num=3642.461):
    for i, palavra in enumerate(lista):
        if palavra == num:
            result = i
    return result
```

```python
df = pd.DataFrame(matriz_trans, columns=['scan', 'julianday', 'descentRate', 'pressure', 'temp', 'temp2', 'cond', 'cond2', 'oxy', 'v', 'v2', 'flag'])
```

```python
b  =posicoes(df.pressure)
```

```python
engine = create_engine('sqlite:///:memory:')
```

```python
df.to_sql('db_dados', engine, index=False)
```

```python
d = sql.execute('SELECT temp, temp2, cond FROM db_dados where scan = 0 or scan = ' + str(b), engine)

return d

```
